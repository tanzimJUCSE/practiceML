```python
import pandas as pd
```

```python
data=pd.read_csv('home_prices.csv')
```

```python
data
```

```python
dummy=pd.get_dummies(data.town)
dummy
```

```python
data_dummy=pd.concat([data,dummy],axis='columns')
data_dummy
```

```python
data_dummy.drop('town',axis='columns',inplace=True)
data_dummy
```

```python
data_dummy.drop('tangail',axis='columns',inplace=True)
data_dummy
```

```python
X=data_dummy.drop('price',axis='columns')
Y=data_dummy.price
X
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
```

```python
model.predict(X)
```

```python
model.score(X,Y)
```

```python
model.predict([[3400,0,0]])#Home price in Tangail
```
